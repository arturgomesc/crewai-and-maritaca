```python
from dotenv import load_dotenv
import maritalk
import os 

load_dotenv()
API_KEY = os.getenv('MARITACA_API_KEY')

model = maritalk.MariTalk(
    key=API_KEY,
    model='sabia-2-small'
)
```

```python
from crewai import Agent, Task, Crew 

gramatica = Agent(
    role='Professor Especialista em Gramática',
    goal='Buscar melhorias gramaticais em um texto',
    backstory=""" Você é um excelente professor de português renomado por ter uma gramática e escrita impecável.
     Sua responsabilidade é corrigir um texto redação e buscar por melhorias gramaticais do início ao fim,
     de acordo com as boas normas da gramática portuguesa.
    """,
    verbose=True
)

semantica = Agent(
    role='Pesquisador Especialista em Semântica',
    goal='Corrigir elementos de semântica de um texto',
    backstory="""Você é um pesquisador da língua portuguesa, muitos dos seus estudos são voltados para a avaliação semântica de diversos textos. 
     Sua responsabilidade é corrigir um texto redação e buscar por melhorias semânticas do início ao fim,
     de acordo com as boas normas da língua portuguesa.
    """,
    verbose=True
)

task = Task(
    expected_output="Uma correção gramatical do seguinte texto: {input}",
    description="Uma correção gramatical do seguinte texto: {input}",
    agent=gramatica
)

task2 = Task(
    expected_output="Uma correção semântica do seguinte texto: {input}",
    description="Uma correção semântica do seguinte texto: {input}",
    agent=semantica
)

my_crew = Crew(
    agents=[gramatica, semantica],
    tasks=[task, task2],
    manager_llm=model
)
crew = my_crew.kickoff(inputs={"input": "O fogo gelado queimou a casa."})
```
